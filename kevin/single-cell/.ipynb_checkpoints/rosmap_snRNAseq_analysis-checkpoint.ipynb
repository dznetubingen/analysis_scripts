```python
import pandas as pd
import numpy as np
import scanpy.api as sc

sc.settings.verbosity = 3
sc.settings.set_figure_params(dpi=150, color_map='viridis')
sc.logging.print_versions()
out_path = "/media/kevin/89a56127-927e-42c0-80de-e8a834dc81e8/rosmap_snRNAseq/"
data_path = "/media/kevin/89a56127-927e-42c0-80de-e8a834dc81e8/rosmap_snRNAseq/data/"
```

# Load data

```python
adata = sc.read(data_path + "filtered_count_matrix.mtx", cache=True).T
genes = list(pd.read_csv(data_path + "filtered_gene_row_names.txt", sep="\t", header=None)[0])
adata.var_names = genes
md = pd.read_csv(data_path + "filtered_column_metadata.txt", sep="\t")
adata.obs_names = md['TAG']
adata
```

```python
# add metadata informatic
adata.obs['projid'] = list(md['projid'])
adata.obs['broadCellType'] = list(md['broad.cell.type'])
adata.obs['subcluster'] = list(md['Subcluster'])
adata.obs
```

```python
sc.pp.filter_cells(adata, min_genes=500)
sc.pp.filter_genes(adata, min_cells=5)
adata.obs['n_counts'] = adata.X.sum(axis=1)
```

```python
sc.pl.scatter(adata, x="n_counts", y='n_genes')
```

```python
# Filter
adata = adata[adata.obs['n_genes'] < 7000, :]
adata = adata[adata.obs['n_counts'] < 20000, :]
adata.raw = sc.pp.log1p(adata, copy=True)
# Normalize per cell
sc.pp.normalize_per_cell(adata)
```

```python
# Save the complete matrix for later
df = pd.DataFrame(adata.X.todense())
df.columns = adata.var.index
#df.to_csv(out_path + "rosmap_norm_counts_all.txt", sep="\t")
```

```python
filter_result = sc.pp.filter_genes_dispersion(adata.X, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.filter_genes_dispersion(filter_result)
```

```python
adata = adata[:, filter_result.gene_subset]
sc.pp.normalize_per_cell(adata)
sc.pp.log1p(adata)
sc.pp.regress_out(adata, ['n_counts'])
sc.pp.scale(adata)
```

```python
sc.pp.neighbors(adata)
sc.tl.umap(adata)
```

```python
sc.tl.louvain(adata, resolution=0.5)
```

```python
sc.pl.umap(adata, color='broadCellType')
```

## Save celltypes

```python
# Change cell type names
celltypes = pd.DataFrame(adata.obs['broadCellType'])
celltypes.broadCellType.replace(['Ast', 'End', 'Ex', 'In', 'Mic', 'Oli', 'Opc', 'Per'],
                               ['Astrocytes', 'Endothelial', 'ExNeurons', 'InNeurons', 'Microglia',
                               'Oligodendrocytes', 'OPC', 'Pericytes'], inplace=True)
adata.obs['celltype'] = celltypes.broadCellType
sc.pl.umap(adata, color='celltype')
```

## Split by subject

```python
out_dir = out_path + "rosmap_snRNAseq_processed/"
subjects = list(set(adata.obs['projid']))
count = 0
for s in subjects:
    count += 1
    print(count)
    print(s)
    keep = list(adata.obs['projid'] == s)
    adata_tmp = adata[keep,]
    df_tmp = df.loc[keep,]
    
    # Save the celltypes
    celltypes = pd.DataFrame(adata_tmp.obs['celltype'])
    celltypes.columns = ['Celltype']
    celltypes.to_csv(out_dir + "rosmap"+str(s)+"_celltypes.txt", sep="\t")
    
    # Save matrix
    df_tmp.to_csv(out_dir + "rosmap"+str(s)+"_norm_counts_all.txt", sep="\t")
    
```
